```python
import pandas as pd
df=pd.read_csv('CAR_DATA.csv')
df
```

```python

dummies=pd.get_dummies(df.CAR_MODEL)
dummies
```

```python
merge=pd.concat ([df, dummies], axis='columns')
merge
```

```python
fore= merge.drop (['CAR_MODEL','MERCEDES'], axis='columns')
fore
```

```python
x= fore.drop (['SELL_PRICE'], axis= 'columns')
x
```

```python
y= fore.SELL_PRICE
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit (x,y)
```

```python
model.predict ([[91000,	8,	True,False]])
```

```python
model.predict ([[60000,	2,	False	,False]])
```

```python
score=model.score (x,y)
score
```

```python
percentage= score/1 *100
percentage
```

```python
print (int(percentage), "%")
```
